```python
import heapq
from operator import itemgetter
from compiler import NATIVE_INSTRUCTIONS
import pyperclip

instructions = [
    "cmp"
]

bigestInstID = heapq.nlargest(1, NATIVE_INSTRUCTIONS.items(), key=itemgetter(1))[0][1] + 1

combinations = [
    "register",
    "registerpointer",
    "number",
    "address"
]

cantStartWith = [
    "number"
]

cantEndWith = [
]

genSubstitudes = {
    "address" : "variable"
}

currInstId = bigestInstID

clipbord = ""

for inst in instructions:
    for val in combinations:
        for val2 in combinations:
            if val in cantStartWith:
                continue
            if val2 in cantEndWith:
                continue
            for k, v in genSubstitudes.items():
                if k == val and k == val2:
                    print(f"\"{inst} %{v}, %{v}\": {currInstId},")
                    clipbord += f"\"{inst} %{v}, %{v}\": {currInstId},\n"
                elif k == val2:
                    print(f"\"{inst} %{val}, %{v}\": {currInstId},")
                    clipbord += f"\"{inst} %{val}, %{v}\": {currInstId},\n"
                elif k == val:
                    print(f"\"{inst} %{v}, %{val2}\": {currInstId},")
                    clipbord += f"\"{inst} %{v}, %{val2}\": {currInstId},\n"

            print(f"\"{inst} %{val}, %{val2}\": {currInstId},")
            clipbord += f"\"{inst} %{val}, %{val2}\": {currInstId},\n"
            currInstId+=1

pyperclip.copy(clipbord)
```

```python
import heapq
from operator import itemgetter
from compiler import NATIVE_INSTRUCTIONS
import pyperclip

instructions = [
    "call"
]

bigestInstID = heapq.nlargest(1, NATIVE_INSTRUCTIONS.items(), key=itemgetter(1))[0][1] + 1

combinations = [
    "label"
]

genSubstitudes = {
    "label" : "number",
    "address" : "variable"
}

currInstId = bigestInstID

clipbord = ""

for inst in instructions:
    for val in combinations:
            for k, v in genSubstitudes.items():
                if k == val:
                    print(f"\"{inst} %{v}\": {currInstId},")
                    clipbord += f"\"{inst} %{v}\": {currInstId},\n"

            print(f"\"{inst} %{val}\": {currInstId},")
            clipbord += f"\"{inst} %{val}\": {currInstId},\n"
            currInstId+=1

pyperclip.copy(clipbord)
```

```python
from compiler import NATIVE_INSTRUCTIONS
import pyperclip

def collapse_dict(input_dict):
    values = set(input_dict.values())
    res = {}

    dictindex = 0

    for value in values:
        dictindex+=1
        for kk, vv in input_dict.items():
            if value == vv:
                res[kk] = dictindex

    return res


clipbord = ""

for k, v in collapse_dict(NATIVE_INSTRUCTIONS).items():
    print(f"\"{k}\": {v},")
    clipbord += f"\"{k}\": {v},\n"

pyperclip.copy(clipbord)
```
